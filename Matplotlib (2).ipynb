<h3>Matplotlib</h3>
<p>Matplotlib offers a variety of plot types, each suited for specific data visualization needs.</p>

<h1>Basic Plotting</h1>

<h3>Line Plot (<code>plt.plot()</code>)</h3>
<p><strong>Purpose:</strong> Displays continuous data trends over an ordered axis (e.g., time).</p>
<p><strong>When to Use:</strong> Ideal for showing changes or trends over time or another continuous variable.</p>
<p><strong>Example:</strong> Tracking daily temperature over a month.</p>
<p><strong>Scenario:</strong> You have temperature data for 30 days, and you want to see how it fluctuates.</p>
<p><strong>Why:</strong> A line plot connects the data points, making it easy to observe trends or patterns.</p>


```python
#syntax: plt.plot(x, y, [fmt], **kwargs)
#example-1
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,20,15,30,90]
plt.plot(x,y)
plt.show()
```

```python
#example-2
plt.plot(x,y, "go--")
plt.show()
```

```python
#example-3
y2=[5,15,20,3,9]
plt.plot(x,y, "b*--", label="line_1")
plt.plot(x,y2, "ro-", label="line_2")
plt.legend()
plt.show()
```

```python
#example-4
plt.plot(x,y, color="purple", linestyle="-", linewidth=2, marker="s", markersize=10)
plt.title("sample plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

<h3>Scatter Plot (<code>plt.scatter()</code>)</h3>
<p><strong>Purpose:</strong> Shows the relationship between two variables using individual points.</p>
<p><strong>When to Use:</strong> Best for visualizing correlations or distributions, especially when data isnt necessarily continuous.</p>
<p><strong>Example:</strong> Comparing students test scores vs. hours studied.</p>
<p><strong>Scenario:</strong> You have data on 50 students with their study hours and test scores.</p>
<p><strong>Why:</strong> Scatter plots reveal if theres a correlation (e.g., more study hours = higher scores) without assuming a continuous trend.</p>


```python
#syntax: plt.scatter(x, y, s=None, c=None, marker=None, **kwargs)
# Example 1: Basic Scatter Plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 15]
plt.scatter(x,y)
plt.show()
```

```python
#Example 2: Custom size and color
plt.scatter(x,y, color='r',s=75)
```

```python
#Example 3: Variable size and color
sizes = [20, 50, 100, 200]
colors = ['blue', 'green', 'red', 'purple']
plt.scatter(x,y, s=sizes, c=colors)
plt.show()
```

```python
#Example 4: With Labels and Transparency
plt.scatter(x, y, s=150, c='orange', alpha=0.5, label="Data Points") #alpha adjusts transparancy
plt.title("scatter plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

<h3>Bar Plot (<code>plt.bar()</code>)</h3>
<p><strong>Purpose:</strong> Compares discrete categories or groups with numerical values.</p>
<p><strong>When to Use:</strong> Useful for categorical data where you want to show quantities or counts.</p>
<p><strong>Example:</strong> Sales of different products in a store.</p>
<p><strong>Scenario:</strong> You have sales data for 5 products (e.g., apples, bananas, etc.).</p>
<p><strong>Why:</strong> Bar plots make it easy to compare the height of bars, highlighting differences across categories.</p>


```python
#plt.bar(x, height,, width=0.8, **kwargs)
# Example 1: Basis Bar Plot
x = ['A', 'B', 'C', 'D']
heights = [10, 20, 15, 25]
plt.bar(x, heights)
plt.show()
```

```python
#Example 2: Custom color and width
plt.bar(x, heights, color='skyblue', width=0.65)
plt.show()
```

```python
#Example 3: With Edge color and labels
plt.bar(x, heights, color='yellow', edgecolor='green', linewidth=0.5)
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Example 4: Stacked Bars
sample_2=[5,10,8,12]
plt.bar(x, sample_2, color='gray', label='Base')
plt.bar(x, heights, bottom=sample_2, color='orange', label='Top')
plt.legend()
plt.show()
```

<h3>Horizontal Bar Plot (<code>plt.barh()</code>)</h3>
<p><strong>Purpose:</strong> Similar to a bar plot but with horizontal bars.</p>
<p><strong>When to Use:</strong> When category names are long or you want a different visual orientation.</p>
<p><strong>Example:</strong> Ranking top 10 countries by population.</p>
<p><strong>Scenario:</strong> You have population data for countries with long names.</p>
<p><strong>Why:</strong> Horizontal bars accommodate longer labels and can be more readable.</p>

```python
#syntax: plt.barh(y, width, height=0.8, **kwargs)

#example 1: Base Horizontal Bar plot
y = ['A', 'B', 'C', 'D']
widths = [10, 20, 15, 25]
plt.barh(y, widths)
plt.show()
```

```python
#Example 2: Custom Colors
plt.barh(y, widths, color='green')
plt.show()
```

```python
#Example 3: with Labels
plt.barh(y, widths, color="teal", height=0.6)
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

<h3>Histogram (<code>plt.hist()</code>)</h3>
<p><strong>Purpose:</strong> Displays the distribution of a single numerical variable.</p>
<p><strong>When to Use:</strong> When you want to see the frequency or count of values within ranges (bins).</p>
<p><strong>Example:</strong> Distribution of exam scores in a class.</p>
<p><strong>Scenario:</strong> You have scores of 100 students and want to see how theyre spread (e.g., most scored 70-80?).</p>
<p><strong>Why:</strong> Histograms show the shape of the data (e.g., normal, skewed).</p>


```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
import numpy as np
data=np.array(data)
data=np.tile(data,2)
print(data)
```

```python
#syntax: plt.hist(x, bins=None, **kwargs)
#example 1: Basix Histogram
plt.hist(data)
plt.show()
```

```python
#example 2: Custom Bins and color
plt.hist(data, bins=5, color='purple', edgecolor='yellow')
plt.show()
```

```python
#example 3: with transparency and labels
plt.hist(data, bins=4, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

<h2>Box Plot (<code>plt.boxplot()</code>)</h2>
<p><strong>Purpose:</strong> Summarizes data distribution (median, quartiles, outliers).</p>
<p><strong>When to Use:</strong> When you need to compare distributions across groups or detect outliers.</p>
<p><strong>Example:</strong> Comparing salaries across different job roles.</p>
<p><strong>Scenario:</strong> You have salary data for engineers, managers, and analysts.</p>
<p><strong>Why:</strong> Box plots quickly show the spread, central tendency, and any extreme values.</p>


```python
#plt.boxplot(x, vert=True, **kwargs)
#example 1:Basis Box Plot
import matplotlib.pyplot as plt
data = [1, 2, 3, 4, 5, 10, 15, 20]
plt.boxplot(data)
plt.show()
```

```python
#Example 2: Multiple Box Plots
data2 = [2, 4, 6, 8, 10, 12]
plt.boxplot([data,data2], labels=['Group1','Group2'])
plt.show()
```

```python
#Example 3: Custom Styling
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("box plot example")
plt.ylabel("Values")
plt.show()
```

<h2>Pie Chart (<code>plt.pie()</code>)</h2>
<p><strong>Purpose:</strong> Shows proportions or percentages of a whole.</p>
<p><strong>When to Use:</strong> When you have a small number of categories and want to visualize their relative contributions.</p>
<p><strong>Example:</strong> Market share of smartphone brands.</p>
<p><strong>Scenario:</strong> You have market share data for 5 brands (e.g., Apple 40%, Samsung 30%, etc.).</p>
<p><strong>Why:</strong> Pie charts emphasize how each category contributes to 100%.</p>


```python
#syntax: plt.pie(x, labels=None, autopct=None, **kwargs)
#Example 1: Basic Pie chart
sizes = [30,40,25,25]
plt.pie(sizes)
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.show()
```

```python
# Example 3: Exploded Slice and Colors
explode = [0, 0.05, 0.1, 0.075]  # Explode the second slice
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, colors=['red', 'blue', 'green', 'yellow'])
plt.title('Pie Chart Example')
plt.show()
```

<h1>Multiple Plots</h1>

```python
import matplotlib.pyplot as plt

# Example 1: Basic 1x2 Subplot Layout
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 10, 20]
plt.subplot(1,2,1) # 1 row, 2 columns, 1st subplot
plt.plot(x,y)
plt.title("plot 1")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(x, y2)
plt.title('Plot 2')
plt.show()
```

```python
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st subplot
plt.plot(x, y1, 'r-')
plt.title('Top Plot')

plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd subplot
plt.plot(x, y2, 'b-')
plt.title('Bottom Plot')

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
# Example 3: 2x2 Grid
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, y1)
plt.title('Top-Left')

plt.subplot(2, 2, 2)  # 2nd subplot
plt.plot(x, y2)
plt.title('Top-Right')

plt.subplot(2, 2, 3)  # 3rd subplot
plt.scatter(x, y1)
plt.title('Bottom-Left')

plt.subplot(2, 2, 4)  # 4th subplot
plt.scatter(x, y2)
plt.title('Bottom-Right')

plt.tight_layout()
plt.show()
```

```python
#fig, ax = plt.subplots(nrows=1, ncols=1, **kwargs)
# Example 1: Basic 1x2 Subplots
fig, ax = plt.subplots(1, 2)  # 1 row, 2 columns
ax[0].plot(x, y1)  # First subplot
ax[0].set_title('Plot 1')
ax[1].plot(x, y2)  # Second subplot
ax[1].set_title('Plot 2')
plt.show()

# Example 2: 2x1 Vertical Subplots
fig, ax = plt.subplots(2, 1)
ax[0].plot(x, y1, 'g-')
ax[0].set_title('Top Plot')
ax[1].plot(x, y2, 'm-')
ax[1].set_title('Bottom Plot')
fig.tight_layout()  # Adjust spacing
plt.show()

# Example 3: 2x2 Grid with Shared Axes
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)  # Share x and y axes
ax[0][0].plot(x, y1)
ax[0, 0].set_title('TL')
ax[0, 1].plot(x, y2)
ax[0, 1].set_title('TR')
ax[1, 0].scatter(x, y1)
ax[1, 0].set_title('BL')
ax[1, 1].scatter(x, y2)
ax[1, 1].set_title('BR')
plt.tight_layout()
plt.savefig('sample.png')
plt.show()
```

<h1>Plot Styling in Matplotlib</h1>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Example 1: Default Style (for comparison)
plt.plot(x, y)
plt.title('Default Style')
plt.show()

# Example 2: 'ggplot' Style
plt.style.use('ggplot')
plt.plot(x, y)
plt.title('ggplot Style')
plt.show()
```

```python
# Example 1: Basic Dashed Line
plt.plot(x, y, '--')  # Dashed line
plt.title('Dashed Line')
plt.show()

# Example 2: Combined with Color and Marker
plt.plot(x, y, 'r--o')  # Red dashed line with circle markers
plt.title('Red Dashed with Circles')
plt.show()

# Example 3: Different Line Styles
plt.plot(x, y, 'b-')    # Blue solid line
plt.plot(x, [5, 15, 10, 20], 'g--')  # Green dashed line
plt.plot(x, [15, 10, 20, 5], 'r:')   # Red dotted line
plt.title('Multiple Line Styles')
plt.show()
```

```python
# Example 1: Basic Color
plt.plot(x, y, color='red')
plt.title('Red Line')
plt.show()

# Example 2: Hex Color
plt.plot(x, y, color='#00FF00')  # Green in hex
plt.title('Hex Green Line')
plt.show()

# Example 3: Multiple Colored Lines
plt.plot(x, y, color='purple', label='Series 1')
plt.plot(x, [5, 15, 10, 20], color='orange', label='Series 2')
plt.legend()
plt.title('Colored Lines')
plt.show()
```

```python
# Example 1: Basic Markers
plt.plot(x, y, marker='o')  # Circle markers
plt.title('Circle Markers')
plt.show()

# Example 2: Different Marker Types
plt.plot(x, y, marker='s', label='Squares')  # Square markers
plt.plot(x, [5, 15, 10, 20], marker='^', label='Triangles')  # Triangle markers
plt.legend()
plt.title('Different Markers')
plt.show()

# Example 3: Marker with Size and Color
plt.plot(x, y, marker='*', markersize=10, markerfacecolor='yellow', markeredgecolor='black')
plt.title('Styled Star Markers')
plt.show()
```

```python
# Example 1: Basic Line Width
plt.plot(x, y, linewidth=2)
plt.title('Thicker Line')
plt.show()

# Example 2: Varying Widths
plt.plot(x, y, linewidth=1, label='Thin')
plt.plot(x, [5, 15, 10, 20], linewidth=3, label='Thick')
plt.legend()
plt.title('Varying Line Widths')
plt.show()

# Example 3: With Color and Marker
plt.plot(x, y, linewidth=4, color='blue', marker='o')
plt.title('Thick Blue Line with Markers')
plt.show()
```

```python
# Example 1: Basic Transparency
plt.plot(x, y, alpha=0.5)
plt.title('Semi-Transparent Line')
plt.show()

# Example 2: Overlapping Lines
plt.plot(x, y, alpha=0.3, linewidth=5, label='Series 1')
plt.plot(x, [5, 15, 10, 20], alpha=0.7, linewidth=5, label='Series 2')
plt.legend()
plt.title('Overlapping Transparent Lines')
plt.show()

# Example 3: With Markers
plt.plot(x, y, marker='o', alpha=0.4, color='red')
plt.title('Transparent Red Markers')
plt.show()
```
